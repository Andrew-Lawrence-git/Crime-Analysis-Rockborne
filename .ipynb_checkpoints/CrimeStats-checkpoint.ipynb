```python
import os
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

# Loading the Data

```python
#Create the data frames for the data from each police force
herts_data = pd.DataFrame(columns=['Crime ID','Month','Reported by','Falls within','Longitude','Latitude','Location','LSOA code','LSOA name','Crime type','Last outcome category','Context'])
kent_data = pd.DataFrame(columns=['Crime ID','Month','Reported by','Falls within','Longitude','Latitude','Location','LSOA code','LSOA name','Crime type','Last outcome category','Context'])
surrey_data = pd.DataFrame(columns=['Crime ID','Month','Reported by','Falls within','Longitude','Latitude','Location','LSOA code','LSOA name','Crime type','Last outcome category','Context'])
essex_data = pd.DataFrame(columns=['Crime ID','Month','Reported by','Falls within','Longitude','Latitude','Location','LSOA code','LSOA name','Crime type','Last outcome category','Context'])

cwd = os.path.abspath('','\\Data')
```

```python
cwd
```

```python
#Make a list of all files in current working directory
list(os.listdir(cwd))
x=0
#Make a list for each police force
herts_list = []
kent_list = []
surrey_list = []
essex_list = []

#Populate lists with the files for each police force
for i in os.listdir(cwd):
    if 'hertfordshire' in os.listdir(cwd)[x]:
        herts_list.append(os.listdir(cwd)[x])
    if 'kent' in os.listdir(cwd)[x]:
        kent_list.append(os.listdir(cwd)[x])
    if 'surrey' in os.listdir(cwd)[x]:
        surrey_list.append(os.listdir(cwd)[x])
    if 'essex' in os.listdir(cwd)[x]:
        essex_list.append(os.listdir(cwd)[x])
    x = x + 1
```

```python
#Combine each .csv into a single dataframe
x=0
for i in herts_list:
    herts_load = pd.read_csv(herts_list[x])
    herts_data = pd.concat([herts_data,herts_load])
    x=x+1
x=0
for i in kent_list:
    kent_load = pd.read_csv(kent_list[x])
    kent_data = pd.concat([kent_data,kent_load])
    x=x+1
x=0
for i in surrey_list:
    surrey_load = pd.read_csv(surrey_list[x])
    surrey_data = pd.concat([surrey_data,surrey_load])
    x=x+1
x=0
for i in essex_list:
    essex_load = pd.read_csv(essex_list[x])
    essex_data = pd.concat([essex_data,essex_load])
    x=x+1
```

```python
#Load the population data for the LSOAs
lsoa_pop22 = pd.read_excel('sapelsoabroadage20222024.xlsx','Mid-2022 LSOA 2021',skiprows=3)

lsoa_pop23 = pd.read_excel('sapelsoabroadage20222024.xlsx','Mid-2023 LSOA 2021',skiprows=3)

lsoa_pop24 = pd.read_excel('sapelsoabroadage20222024.xlsx','Mid-2024 LSOA 2021',skiprows=3)

lsoa_pop_median = pd.read_excel('sapelsoabroadage20222024.xlsx','Median age LSOA 2011-2024',skiprows=3)
```

```python
#Load the population data for the PFAs
pfa_pop = pd.read_excel('policeforceareas1991to2024.xlsx','Mid-2021 to Mid-2024',skiprows=3)
```

# Data Cleaning

```python
#Split years and months into two columns
#essex
essex_split = essex_data['Month'].str.split('-',expand=True)
essex_split.columns = ['Year','Month']

essex_ym = essex_data.drop('Month',axis=1)
essex_ym.insert(1,'Month',essex_split['Month'])
essex_ym.insert(1,'Year',essex_split['Year'])

#hertfordshire
herts_split = herts_data['Month'].str.split('-',expand=True)
herts_split.columns = ['Year','Month']

herts_ym = herts_data.drop('Month',axis=1)
herts_ym.insert(1,'Month',herts_split['Month'])
herts_ym.insert(1,'Year',herts_split['Year'])

#kent
kent_split = kent_data['Month'].str.split('-',expand=True)
kent_split.columns = ['Year','Month']

kent_ym = kent_data.drop('Month',axis=1)
kent_ym.insert(1,'Month',kent_split['Month'])
kent_ym.insert(1,'Year',kent_split['Year'])

#surrey
surrey_split = surrey_data['Month'].str.split('-',expand=True)
surrey_split.columns = ['Year','Month']

surrey_ym = surrey_data.drop('Month',axis=1)
surrey_ym.insert(1,'Month',surrey_split['Month'])
surrey_ym.insert(1,'Year',surrey_split['Year'])
```

```python
#merge population of each LSOA code with crime data
#hertfordshire
herts_data22 = herts_ym[herts_ym['Year']=='2022']
herts_lsoa22 = pd.merge(herts_data22,lsoa_pop22, left_on=['LSOA code'],right_on=['LSOA 2021 Code'])
```

```python
#Remove rows from the PFA data that aren't necessary (unrelated to the areas being looked at)
pfa_pop_clean = pfa_pop[(pfa_pop['PFA 2023 Name']=='Essex')|(pfa_pop['PFA 2023 Name']=='Hertfordshire')|(pfa_pop['PFA 2023 Name']=='Kent')|(pfa_pop['PFA 2023 Name']=='Surrey')]
pfa_pop_clean.reset_index(drop=True)
```
